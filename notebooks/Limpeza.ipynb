## Tratamento Clientes

```python
#df_clientes["location"].value_counts()

ufs = [
    'AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 
    'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 
    'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO'
]
#df_clientes["location"] = df_clientes["location"].strip()
#df_clientes["location"].head()
```

```python
local = df_clientes["location"]
```

```python
def extrair_estado_cidade(local):
    if not isinstance(local, str):
        return None, None
    
    local = local.strip()
    
    # 🔹 1. Verifica início
    possivel_uf = local[:2]
    
    if possivel_uf in ufs:
        cidade = local[2:].strip(" -_,/")
        return possivel_uf, cidade
    
    # 🔹 2. Verifica final
    possivel_uf = local[-2:]
    
    if possivel_uf in ufs:
        cidade = local[:-2].strip(" -_,/")
        return possivel_uf, cidade
    
    # 🔹 3. Caso não encontre
    return None, local
```

```python
df_clientes[['estado', 'cidade']] = df_clientes["location"].apply(
    lambda x: pd.Series(extrair_estado_cidade(x))
)

df_clientes.head()
```

## Tratamento custos importação


```python
dados_historicos = df_custos_importacao["historic_data"]

df_dados_historicos_explodidos = dados_historicos.explode("historic_data")
df_final = pd.json_normalize(df_dados_historicos_explodidos)
df_custos_importacao = df_custos_importacao.drop(columns=['historic_data'])
df_custos_importacao = pd.concat([df_custos_importacao, df_final], axis = 1)
df_custos_importacao.tail()
```
